```python
with open('file.txt',newline='') as data:
    print(data.read())
```

```python
with open('two.txt','w',newline='') as data:
    data.write("This is Line 1")
    data.write('\n')
    data.write("This is Line 2")
```

```python
with open("demo.txt",'w') as data:
    data.writelines(['I am a boy\n', 'I stay in Mumbai\n','I love coding'])
```

```python
with open('file.txt','w') as test:
    print(test.writable())
```

```python
with open('file.txt','r') as test:
    print(test.read())
```

```python
with open('file.txt','r') as test:
    print(test.read(23))
    print(test.tell())
    test.seek(44)
    print("Current Position of cursor is",test.tell())
```

# copying data from one file to another

```python
with open('demo.txt','r') as data:
    with open('copy.txt','w') as copied:
        for line in data:
            copied.write(line)
```

# append mode in python

```python
with open('city.txt',"w") as data:
    data.writelines(["Delhi is the capital of India\n","Paris is the capital of France\n","Tokyo is the capital of Japan\n"])
```

```python
with open('city.txt','a') as data:
    data.write("Washington DC is the capital of USA \n")
```

```python
with open('hi.txt','a') as data:
    data.write("Test")
```

# file access in binary mode

```python
with open('test.png','rb') as data:
    print(data.read())
```

# copying / writing binary files

```python
with open('test.png','rb') as data:
    with open('copy.png','wb') as copied:
        for line in data:
            copied.write(line)
```
